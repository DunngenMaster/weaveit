```python
# Import required libraries
import os
import google.generativeai as genai

# Set your Gemini API key
GEMINI_API_KEY = "your-gemini-api-key-here"  # Replace with your actual key

# Configure Gemini
genai.configure(api_key=GEMINI_API_KEY)
```

```python
# List available models
print("Available Gemini models:\n")
available_models = []

for m in genai.list_models():
    if 'generateContent' in m.supported_generation_methods:
        available_models.append(m.name)
        print(f"âœ“ {m.name}")

print(f"\nTotal models: {len(available_models)}")
```

```python
# Initialize the model
# Using gemini-2.5-flash (fast and efficient)
model_name = 'gemini-2.5-flash'
model = genai.GenerativeModel(model_name)

print(f"Using model: {model_name}")
print("Model initialized successfully!")
```

```python
# Test single message
response = model.generate_content("Hello! Introduce yourself in one sentence.")

print("User: Hello! Introduce yourself in one sentence.")
print(f"Gemini: {response.text}")
```

```python
# Interactive chat session
def start_chat():
    print("Chat started! Type 'quit' to exit.\n")
    
    # Start chat session
    chat = model.start_chat(history=[])
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() in ['quit', 'exit', 'bye']:
            print("\nGoodbye!")
            break
        
        if not user_input.strip():
            continue
        
        response = chat.send_message(user_input)
        print(f"Gemini: {response.text}\n")

# Uncomment to start interactive chat
# start_chat()
```
